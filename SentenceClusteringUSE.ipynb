**Prepare Data**

```python
# Step 1: Install required packages
!pip install tensorflow tensorflow-hub

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 2: Import libraries
import tensorflow as tf
import tensorflow_hub as hub

# Step 3: Load the Universal Sentence Encoder from TF Hub
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
print("USE model loaded!")

# Step 4: Define your sentences
```

```python
sentences = [
    "I love machine learning.",
    "I like machine learning",
    "machine learning is a great subject to learn",
    "machine learning is my favorite",
    "machine learning is fantastic",
    "machine learning is wonderful",
    "The weather is sunny today.",
    "Python is a great programming language.",
    "I love programming in python",
    "How are you doing?",
    "Google Colab makes it easy to use GPUs for free."
]

# Step 5: Generate sentence embeddings
embeddings = embed(sentences)

# Step 6: Display the embeddings
import numpy as np

for i, sentence in enumerate(sentences):
    print(f"\nSentence: {sentence}")
    print(f"Embedding shape: {embeddings[i].shape}")
    print(f"Embedding preview (first 5 values): {np.round(embeddings[i][:5].numpy(), 4)}")
# Step 9: Convert embeddings to NumPy and normalize
X = embeddings.numpy()
X_scaled = StandardScaler().fit_transform(X)

# Step 10: Run DBSCAN
# You may need to tune `eps` and `min_samples` based on your data
db = DBSCAN(eps=0.5, min_samples=2, metric='cosine').fit(X_scaled)
labels = db.labels_

# Step 11: Print results
from collections import defaultdict

clusters = defaultdict(list)
for sentence, label in zip(sentences, labels):
    clusters[label].append(sentence)

print("\n=== DBSCAN Clustering Results ===")
for label, cluster_sentences in clusters.items():
    cluster_name = f"Cluster {label}" if label != -1 else "Noise"
    print(f"\n{cluster_name}:")
    for s in cluster_sentences:
        print(f" - {s}")
```
